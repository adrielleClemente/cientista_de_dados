```python
y, X = patsy.dmatrices('np.log(renda) ~ C(sexo) + C(posse_de_veiculo) + C(posse_de_imovel) + qtd_filhos + C(tipo_renda) + C(educacao) + C(estado_civil) + C(tipo_residencia) + idade + tempo_emprego + qt_pessoas_residencia', df)
X
```

```python
sm.OLS(y, X).fit().summary()
```

```python
y, X = patsy.dmatrices('np.log(renda) ~ C(tipo_renda, Treatment("Servidor público")) + C(estado_civil, Treatment("Solteiro")) + idade + qt_pessoas_residencia + qtd_filhos', df)
sm.OLS(y, X).fit().summary()
```

**Análise:**

Depois de observar os parâmetros, tirar variáveis e aplicar a casela de referência, precisei voltar com algumas variáveis insignificantes pois o R² abaixou muito. No processo, algumas dummies melhoraram seu p-value, mas nada significativo. Por último, o AIC ficou bem alto nesse último modelo. O primeiro modelo gerado, sem muito tratamento se saiu melhor do que o final.
