````python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

from sklearn.tree import plot_tree
from sklearn.tree import DecisionTreeClassifier

from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix
from sklearn.metrics import ConfusionMatrixDisplay

from sklearn.model_selection import train_test_split
from sklearn.model_selection import cross_val_score
````

````python
df = pd.read_csv('demo_dummie.csv')
df.head()
````

````python
# verificando dados missings
df.isna().sum()
````

````python
# verificando tipo de dados
df.dtypes
````

````python
df['mau'].value_counts()
````

````python
# transformando o campo bool em int
df['mau'] = df['mau'].astype(int)
````

````python
df['mau'].value_counts()
````

````python
y = df.mau
y
````

````python
x = df.drop('mau', axis=1)
x.head(3)
````

````python
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size = 0.3, random_state = 100)
````

````python
clf = DecisionTreeClassifier(random_state=100)
clf = clf.fit(x_train,  y_train)
````

````python
plt.figure(figsize=[25,25])
plot_tree(clf,
         filled=True,
         rounded=True,
         class_names=['bom', 'mau'],
         feature_names = x_train.columns);
````

````python
cm = confusion_matrix(y_train, clf.predict(x_train), labels=clf.classes_)
MatrizDeConfusao = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=['Aprovado', 'Reprovado'])

MatrizDeConfusao.plot();
````

````python
predict = clf.predict(x_test)
````

````python
accuracy_score(y_test, predict)
````

````python
y_chapeu_teste = clf.predict(x_test)
````

````python
cm = confusion_matrix(y_test, clf.predict(x_test), labels=clf.classes_)
MatrizDeConfusao = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=['Aprovados', 'Reprovados'])

MatrizDeConfusao.plot();
````

````python
accuracy_score(y_test, y_chapeu_teste)
````

````python
# Treine uma nova árvore com número mínimo de observações por folha de 5 e máximo de profundidade de 10. Use o random_state = 123.
````

````python
clf_nova = DecisionTreeClassifier(max_depth=10, min_samples_leaf=5, random_state=123)
clf_nova = clf_nova.fit(x_train,y_train)
````

````python
cm = confusion_matrix(y_test, clf_nova.predict(x_test), labels=clf.classes_)
MatrizDeConfusao = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=['Aprovados', 'Reprovados'])

MatrizDeConfusao.plot();
````

````python
predict = clf_nova.predict(x_test)
accuracy_score(y_test,predict)
````
