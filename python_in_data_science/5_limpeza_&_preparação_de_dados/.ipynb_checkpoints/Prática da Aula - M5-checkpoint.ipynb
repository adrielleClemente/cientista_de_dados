### Aula 1: Identificando e tratando dados ausentes

Espaço do índice

```python
import pandas as pd
import numpy as np
```

```python
#criando dataframe aleatório
df = pd.DataFrame(np.random.randn(9,4)*100,
                 index=['A','B','C','D','E','F','G','H','i'],
                 columns=['Coluna1','Coluna2','Coluna3','Coluna4'])
df
```

```python
df.dtypes
```

#### Alterando o tipo da coluna

```python
df['Coluna1']=df['Coluna1'].astype(int)
df['Coluna1']
df.dtypes
```

```python
df['Coluna3']=df['Coluna3'].astype(str)
df['Coluna3']
df.dtypes
```

#### Simulando Dados Faltantes para o Tratamento

```python
# Acesso da informação
df.iloc[4,2]
```

```python
# Inserindo o NaN
df.iloc[4,2] = np.nan
df
```

```python
df.iloc[1,0] = np.nan
df.iloc[4,0] = np.nan
df.iloc[3,0] = np.nan
df.iloc[8,0] = np.nan
df.iloc[6,0] = np.nan
df.iloc[4,3] = np.nan
df.iloc[4,3] = np.nan
df.iloc[8,3] = np.nan
df
```

#### Identificando Dados Ausentes

```python
# mesma coisa `df.isnull()`
df.isna()
```

```python
# Matriz booleana
df['Coluna3'].isna()
```

```python
# Retorna os únicos valores NaN da coluna selecionada.(por consequencia, retorna valores de outras colunas na mesma linha.)
df[df['Coluna3'].isna()]
```

```python
# False: Retorna valores não NaN.
df[~df['Coluna3'].isna()]
```

```python
# soma dos NaN
df.isna().sum()
```

```python
# Porcentagem
df.isna().sum()/len(df)*100
```

#### Tratando Dados Faltantes

```python
# Preenchimeto por 0 (substituição)
df['Coluna1'].fillna(0)
```

```python
df
```

```python
# Média
media = df['Coluna4'].mean()
df['Coluna4'].fillna(media)
```
