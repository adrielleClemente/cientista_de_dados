### I. Modelo no valor da gorjeta

1. Crie a matriz de design (e a matriz y) utilizando o Patsy, para um modelo em ```tip```, explicada por ```sex, smoker, diner e net_bill```.  
2. Remova as variáveis não significantes.  
3. observe o gráfico de resíduos em função de ```net_bill```  
4. teste transformar ```net_bill``` no log e um polinômio. Escolha o melhor modelo.

````python
import pandas as pd
import seaborn as sns
from seaborn import load_dataset
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

import patsy
import statsmodels.api as sm
import statsmodels.formula.api as smf

%matplotlib inline
````

````python
tips = load_dataset("tips")
tips['tip_pct'] = tips['tip'] / (tips['total_bill'] - tips['tip'])
tips['net_bill'] = tips['total_bill'] - tips['tip']
tips.head()
````

````python
y, X = patsy.dmatrices('tip ~ sex + smoker + time + net_bill + 1', tips)
````

````python
modelo = sm.OLS(y,X).fit()

modelo.summary()
````

````python
# removendo variáveis
y, X = patsy.dmatrices('tip ~ smoker + net_bill + 1', tips)
````

````python
reg = sm.OLS(y,X).fit()

modelo.summary()
````

````python
tips['res_log'] = reg.resid
sns.scatterplot(x = 'net_bill', y = 'res_log', data = tips)
plt.axhline(y=0, color='r', linestyle='--')
````

````python
# log de net_bill
reg = smf.ols('tip ~ smoker + np.log(net_bill) +1', data = tips).fit()
reg.summary()
````

````python
# polinômio de 2° grau
reg = smf.ols('tip ~ smoker + np.power(net_bill,2) +1', data = tips).fit()
reg.summary()
````

**Escolhendo o Melhor**

Entre o modelo que tem net_bill no log e o modelo ajustado com polinômio de 2° grau, este último se sai melhor por exibir o R-quadrado no valor de 0.304, enquanto o log mostra o valor de 0.295.

### II. Modelo no valor do percentual da gorjeta

1. Crie a matriz de design (e a matriz y) utilizando o Patsy, para um modelo no log de ```tip```, explicado por ```sex, smoker, diner e net_bill```.
2. Remova as variáveis não significantes.
3. Observe o gráfico de resíduos em função de ```net_bill```
4. Teste transformar ```net_bill``` no log e um polinômio. Escolha o melhor modelo.
5. Do modelo final deste item, calcule o $R^2$ na escala de ```tip``` (sem o log). Compare com o modelo do item 1. Qual tem melhor coeficiente de determinação?

````python
y, X = patsy.dmatrices('np.log(tip) ~ sex + smoker + time + net_bill + 1', tips)
modelo = sm.OLS(y,X).fit()

modelo.summary()
````

````python
# removendo variáveis
y, X = patsy.dmatrices('np.log(tip) ~ smoker + net_bill + 1', tips)
modelo = sm.OLS(y,X).fit()

modelo.summary()
````

````python
tips['res_log'] = modelo.resid
sns.scatterplot(x = 'net_bill', y = 'res_log', data = tips)
plt.axhline(y=0, color='r', linestyle='--')
````

````python
y, X = patsy.dmatrices('np.log(tip) ~ smoker + np.log(net_bill) + 1', tips)
modelo = sm.OLS(y,X).fit()

modelo.summary()
````

````python
y, X = patsy.dmatrices('np.log(tip) ~ smoker + np.power(net_bill,2) + 1', tips)
modelo = sm.OLS(y,X).fit()

modelo.summary()
````

**Escolhendo o Melhor**

Para esse teste, o melhor ajuste se deu com net_bill no log exibindo o valor de 0.319. Comparando com o melhor modelo do exercício anterior, com R² de 0.304, o que se sai melhor é o ajuste é o escolhido deste exercício.
