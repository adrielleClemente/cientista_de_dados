1) Com o que você viu em aula, faça uma análise das variáveis qualitativas (posse de imóvel e posse de veículo) para verificar se são boas preditoras de *renda*. Alguma delas parece discriminar *renda* de forma estatisticamente significante?

````python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
````

````python
df = pd.read_csv('previsao_de_renda.csv')
````

````python
fig, ax = plt.subplots(1, 2, figsize=(10, 5))
sns.countplot(x='posse_de_imovel', data=df, ax=ax[0])
sns.countplot(x='posse_de_veiculo', data=df, ax=ax[1])
plt.show()
````

````python
media_posse_de_imovel = df.groupby('posse_de_imovel')['renda'].mean()
media_posse_de_imovel
````

````python
media_posse_de_veiculo = df.groupby('posse_de_veiculo')['renda'].mean()
media_posse_de_veiculo
````


## Conclusão
O que notei é que a posse de imóvel está positivamente correlacionada com a renda, enquanto a posse de veículo está negativamente correlacionada com a renda.
