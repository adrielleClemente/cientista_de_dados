```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
df = pd.read_csv('Estrutura_SINASC_para_CD.csv')
```

2) Monte um gráficos de barras com as variáveis APGAR1, APGAR5, SEXO, GESTACAO, ESTCIVMAE.

```python
valores = df['APGAR1'].value_counts()
valores = valores.sort_index()
plt.bar(valores.index, valores.values)

plt.xlabel('Valores presentes na coluna APGAR1')
plt.ylabel('Qtde de vezes que os valores aparecem')
plt.title('Coluna APGAR1')
plt.show()
```

```python
valores = df['APGAR5'].value_counts()
valores = valores.sort_index()
plt.bar(valores.index, valores.values)

plt.xlabel('Valores presentes na coluna APGAR5')
plt.ylabel('Qtde de vezes que os valores aparecem')
plt.title('Coluna APGAR5')
plt.show()
```

```python
valores = df['SEXO'].value_counts()
valores = valores.sort_index()
plt.bar(valores.index, valores.values)

plt.xlabel('Valores presentes na coluna SEXO')
plt.ylabel('Qtde de vezes que os valores aparecem')
plt.title('Coluna SEXO')
plt.show()
```

```python
valores = df['GESTACAO'].value_counts()
valores = valores.sort_index()
plt.bar(valores.index, valores.values)

plt.xlabel('Valores presentes na coluna GESTACAO')
plt.xticks(rotation=45)
plt.ylabel('Qtde de vezes que os valores aparecem')
plt.title('Coluna GESTACAO')

plt.show()
```

```python
valores = df['ESTCIVMAE'].value_counts()
valores = valores.sort_index()
plt.bar(valores.index, valores.values)

plt.xlabel('Valores presentes na coluna ESTCIVMAE')
plt.xticks([1,2,3,4,5,9])

plt.ylabel('Qtde de vezes que os valores aparecem')
plt.title('Coluna ESTCIVMAE')

plt.show()
```

#### 3) Monte um gráfico de barras mostrando o APGAR5 médio por CONSULTAS

```python

valores = df['CONSULTAS'].value_counts().sort_index()
# valores = valores.sort_index()
plt.bar(valores.index, valores.values)

plt.xlabel('Valores presentes na coluna CONSULTAS')
plt.xticks([1,2,3,4,9])

plt.ylabel('Qtde de vezes que os valores aparecem')
plt.yticks([51,986,1379,5988,18624])

plt.title('Coluna CONSULTAS')

plt.show()
```

4) Monte um painel com 4 gráficos mostrando as frequencias de SEXO, GESTACAO, ESTCIVMAE e CONSULTAS

```python
plt.close('all')
plt.rc('figure', figsize=(12, 18))
fig, axes = plt.subplots(2, 2)

sns.countplot(ax = axes[0, 0], x='SEXO', data=df)
sns.countplot(ax = axes[0, 1], x='GESTACAO', data=df)
axes[0, 1].set_xticklabels(axes[0, 1].get_xticklabels(), rotation=30, ha="right")
sns.countplot(ax = axes[1, 0], x='ESTCIVMAE', data=df)
sns.countplot(ax = axes[1, 1], x='CONSULTAS', data=df)
```
